### # **Problem:**

- Compare the **Performance** (time taken) of **C, C++, Java, R, Python** Programs for
### Covert 200MB, 400MB, 600MB, 800MB and 1000MB text files to upper case.


image.png

```python
# 1. Import required libraries
import matplotlib.pyplot as plt

# 2. Sample data from the table
file_sizes = [200, 400, 600, 800, 1000]  # in MB

time_taken = {
    "C": [12, 20, 34, 45, 55],
    "C++": [15, 25, 30, 50, 60],
    "Java": [18, 30, 40, 60, 70],
    "R": [20, 35, 45, 60, 80],
    "Python": [25, 40, 53, 75, 100]
}

# 3. Plot the performance graph
plt.figure(figsize=(10, 6))
for lang, times in time_taken.items():
    plt.plot(file_sizes, times, marker='o', label=lang)

# 4. Graph details
plt.title("Text File Conversion to Uppercase - Time Comparison")
plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (seconds)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
